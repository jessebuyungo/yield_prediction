```python
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split( x_1, y_1, test_size=0.33, random_state=42)
```

```python
X_test
```

```python
from sklearn import linear_model
import statsmodels.api as sm
```

```python
regr = linear_model.LinearRegression()
regr.fit(X_train, y_train)
```

```python
print('Intercept: \n', regr.intercept_)
print('Coefficients: \n', regr.coef_)
```

```python
# with statsmodels
x = sm.add_constant(X_train) # adding a constant
x_test = sm.add_constant(X_test)
model = sm.OLS(y_train, x).fit()
predictions = model.predict(x_test)

print_model = model.summary()
print(print_model)
```

```python
#predictions
print('Test_data, Prediction')

for i in range (len(y_test)):
   d_2 = round((np.array(y_test)[i])*1000,2)
   d_1 = round((np.array(predictions)[i])*1000,2)


   print(d_2,', ',d_1)
```

```python

from sklearn.metrics import mean_squared_error
import math
```

```python
MSE = mean_squared_error(y_test*1000,predictions*1000)
RMSE = math.sqrt(MSE)
```

```python
RMSE
```

```python
from sklearn.metrics import r2_score
```

```python
Rsq= r2_score(y_test*1000,predictions*1000)
```

```python
from sklearn.metrics import mean_absolute_error
```

```python
MAE =mean_absolute_error(y_test*1000,predictions*1000)
```

```python
from sklearn.metrics import max_error, mean_absolute_percentage_error
ME= max_error(y_test*1000,predictions*1000)
```

```python
MAP= mean_absolute_percentage_error(y_test*1000,predictions*1000)
```

```python
print('Rsquared_Value:', Rsq)
print('Mean Absolute Error:', MAE)
print('Root Mean Square Value:', RMSE)
print('Max Error:', ME)
print('Mean Absolute percentage Error:', MAP)
```
